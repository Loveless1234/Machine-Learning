```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from pandas_datareader import data,wb
```

```python
from datetime import datetime
```

```python
start = datetime(2006,1,1)
end = datetime(2016,1,1)
```

```python
# Bank of America
BAC = data.DataReader("BAC", 'yahoo', start, end)

# Bank of citiGroup
C = data.DataReader("C", 'yahoo', start, end)

# Bank of goldman sachs
GS= data.DataReader("GS", 'yahoo', start, end)

# Bank of JPMorgan Chase
GPM = data.DataReader("JPM", 'yahoo', start, end)

# Bank of Morgan Stanley
MS = data.DataReader("MS", 'yahoo', start, end)

# Bank of Wells Fargo
WFC = data.DataReader("WFC", 'yahoo', start, end)
```

```python
tickers = ['BAC', 'C', 'GS', 'GPM', 'MS', 'WFC']
```

```python
bank_stocks = pd.concat([BAC, C, GS, GPM, MS, WFC],axis=1,keys=tickers)
```

```python
bank_stocks.head(5)
```

```python
bank_stocks.columns.names = ['Bank Ticker', 'Stock Info']
```

```python
bank_stocks.head()
```

```python
for i in tickers:
    print(i,bank_stocks[i]['Close'].max())
    

    

#bank_stocks['BAC']['Close'].max()
```

```python
bank_stocks.xs(key='Close', axis=1, level='Stock Info').max()
```

```python
returns = pd.DataFrame()
```

```python
for i in tickers:
    returns[i +'Returns']= bank_stocks[i]['Close'].pct_change()
```
