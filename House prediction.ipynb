```python
df=df.drop(['id','date','zipcode','day'],axis=1)
```

```python
df.shape
```

```python
X=df.drop('price',axis=1).values
y=df['price'].values
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train= scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))

model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(X_train, y_train, validation_data=(X_test,y_test),batch_size=128,epochs=400)
```

```python
loss_data = pd.DataFrame(model.history.history)
```

```python
loss_data.plot()
```

```python
prediction = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,explained_variance_score
```

```python
mean_absolute_error(y_test,prediction)
```

```python
mean_squared_error(y_test,prediction)**0.5
```

```python
explained_variance_score(y_test,prediction)
```

```python
plt.figure(figsize=(12,8))
plt.scatter(y_test,prediction)
plt.plot(y_test,y_test,'r')
```
